```python
#Soal Latihan 1: Antrian Pemesanan Tiket Bioskop
class QueueList:
    def __init__(self, kapasitas_max=10):
        self.queue = []
        self.kapasitas_max = kapasitas_max
        self.waiting_list = []  # Antrian tambahan untuk pelanggan yang menunggu

    def enqueue(self, id_tiket):
        # mengecek apakah ini pelanggan VIP atau bukan dengan (kode mengandung angka 7)
        if '7' in id_tiket:
            print(f"yuhuuu vip {id_tiket} dapat kursi tanpa mengantri!")
            return True

        # kalau antrian belum penuh, maka tambahkan pelanggan
        if len(self.queue) < self.kapasitas_max:
            self.queue.append(id_tiket)
            print(f"hai pelanggan {id_tiket} kamu ditambahkan ke antrian. Posisi antrian kamu: {len(self.queue)}")
            return True
        else:
            # kalau antrian penuh, masukkan ke daftar tunggu
            self.waiting_list.append(id_tiket)
            print(f"maaf ya antri, penuh soalnya {id_tiket} tenang kamu ditambahkan ke daftar tunggu. Posisi tunggu kamu: {len(self.waiting_list)}")
            return False

    def dequeue(self):
        if not self.is_empty():
            served_customer = self.queue.pop(0)
            print(f"haloo Pelanggankuu {served_customer} kamu sedang dilayani nieh.")

            # kalau ada pelanggan di daftar tunggu, pindahkan ke antrian tunggu utama
            if len(self.waiting_list) > 0:
                pelanggan_selanjutnya = self.waiting_list.pop(0)
                self.queue.append(pelanggan_selanjutnya)
                print(f"kamu {pelanggan_selanjutnya} dipindahkan dari daftar tunggu ke antrian tunggu utama.")
            return served_customer
        return None

    def front(self):
        return self.queue[0] if not self.is_empty() else None

    def rear(self):
        return self.queue[-1] if not self.is_empty() else None

    def is_empty(self):
        return len(self.queue) == 0

    def size(self):
        return len(self.queue)

    def display_queue(self):
        if self.is_empty():
            print("Antrian kosong nieh.")
        else:
            print(f"Antrian sekarang jumlahnya ({len(self.queue)}/{self.kapasitas_max} pelanggan):")
            for i, customer in enumerate(self.queue, 1):
                print(f"{i}. {customer}")

        if len(self.waiting_list) > 0:
            print(f"\ndaftar tunggu ({len(self.waiting_list)} pelanggan):")
            for i, customer in enumerate(self.waiting_list, 1):
                print(f"{i}. {customer}")


# Fungsi untuk membuat kode pelanggan
def generate_customer_code(nama, npm):
    # Mengambil dua huruf pertama nama
    nama_prefix = nama[:2].upper()
    # Mengambil dua digit terakhir NPM
    npm_suffix = npm[-2:]

    return f"{nama_prefix}{npm_suffix}"


# Program utama
def main():
    bioskop_queue = QueueList(kapasitas_max=10)

    while True:
        print("\nAntrian bioskop")
        print("1. menambahkan pelanggan ke antrian")
        print("2. melayani pelanggan")
        print("3. melihat pelanggan pertama")
        print("4. melihat pelanggan terakhir")
        print("5. mengecek jumlah pelanggan dalam antrian")
        print("6. menampilkan semua antrian")
        print("0. keluar skrng?")

        pilih = input("silahkan pilih menu (0-6): ")

        if pilih == '1':
            nama = input("masukkan nama pelanggan: ")
            npm = input("masukkan NPM pelanggan: ")
            customer_code = generate_customer_code(nama, npm)
            bioskop_queue.enqueue(customer_code)

        elif pilih == '2':
            served = bioskop_queue.dequeue()
            if served is None:
                print("gak ada pelanggan dalam antrian.")

        elif pilih == '3':
            front_customer = bioskop_queue.front()
            if front_customer:
                print(f"pelanggan pertama dalam antrian nih: {front_customer}")
            else:
                print("gak ada pelanggan dalam antrian.")

        elif pilih == '4':
            rear_customer = bioskop_queue.rear()
            if rear_customer:
                print(f"pelanggan terakhir dalam antrian nih: {rear_customer}")
            else:
                print("gak ada pelanggan dalam antrian.")

        elif pilih == '5':
            print(f"jumlah pelanggan dalam antrian ada: {bioskop_queue.size()}")

        elif pilih == '6':
            bioskop_queue.display_queue()

        elif pilih == '0':
            print("bye guys have fun!")
            break

        else:
            print("pilih yang valid dong, coba ulangi lagi.")


# untuk menjalankan program menggunakan main
main()
```

```python
#Soal Latihan 2 Antrian Layanan Fotokopi Kampus
class QueueList:
    def __init__(self, kapasitas_max=20):
        self.queue = []
        self.kapasitas_max = kapasitas_max

    def enqueue(self, order_id, num_pages):
        # memeriksa apakah ini pesanan cepat kilat (kurang dari 10 halaman)
        if num_pages < 10:
            print(f"yuhuuu cepat kilat! Pesanan {order_id} dengan {num_pages} halaman langsung kita proses!")
            return True

        # menambahkan pesanan ke antrian
        self.queue.append({"order_id": order_id, "num_pages": num_pages, "processed": 0})
        print(f"Pesanan fotokopi {order_id} dengan {num_pages} halaman ditambahkan ke antrian. urutan ke: {len(self.queue)}")
        return True

    def dequeue(self):
        if not self.is_empty():
            current_order = self.queue[0]

            # kalau pesanan lebih dari 100 halaman dan kebetulan belum diproses setengahnya maka,
            if current_order["num_pages"] > 100 and current_order["processed"] == 0:
                # memproses sesi pertama (setengah)
                current_order["processed"] = 1

                # memindahkan antriab ke belakang
                self.queue.append(self.queue.pop(0))

                print(f"udah selesai nih sesi 1 {current_order['order_id']}yuhuu ({current_order['num_pages']//2} halaman). Memindahkan ke belakang antrian untuk sesi yang ke 2.")
                return None
            else:
                # kalau udah sesi 2 atau kurang dari 100 halaman, maka selesaikan pesanan
                served_order = self.queue.pop(0)
                print(f"Pesanan {served_order['order_id']} dengan {served_order['num_pages']} halaman sudah selesai kami proses.")
                return served_order
        return None

    def front(self):
        return self.queue[0] if not self.is_empty() else None

    def rear(self):
        return self.queue[-1] if not self.is_empty() else None

    def is_empty(self):
        return len(self.queue) == 0

    def size(self):
        return len(self.queue)

    def display_queue(self):
        if self.is_empty():
            print("wow kosong.")
        else:
            print(f"antrian sekarang ({len(self.queue)} pesanan):")
            for i, order in enumerate(self.queue, 1):
                status = "sesi 2" if order["processed"] == 1 else "maaf belum diproses"
                print(f"{i}. dengan ID: {order['order_id']}, halaman: {order['num_pages']}, status: {status}")


# membuat ID pemesanan
def generate_order_id(nama, npm):
    # mengambil tiga huruf pertama nama
    nama_prefix = nama[:3].upper()
    # mengambil dua digit terakhir NPM
    npm_suffix = npm[-2:]

    return f"{nama_prefix}{npm_suffix}"


# Program utama
def main():
    fotokopi_queue = QueueList()

    while True:
        print("\nlayanan fotokopi kampuss")
        print("1. menambah pesanan ke antrian")
        print("2. memproses pesanan")
        print("3. melihat pesanan pertama")
        print("4. melihat pesanan terakhir")
        print("5. mengecek jumlah pesanan antrian")
        print("6. menampilkan Semua Pesanan")
        print("0. Keluar")

        pilih = input("silahkan pilih menu (0-6): ")

        if pilih == '1':
            nama = input("masukkan nama kamu(mahasiswa): ")
            npm = input("masukkan NPM kamu(mahasiswa): ")
            order_id = generate_order_id(nama, npm)

            try:
                num_pages = int(input("masukkan jumlah halamannya: "))
                if num_pages <= 0:
                    print("jumlah halaman harus lebih dari 0 ya bub")
                    continue
                fotokopi_queue.enqueue(order_id, num_pages)
            except ValueError:
                print("jumlah halaman harus banget berupa angka")

        elif pilih == '2':
            served = fotokopi_queue.dequeue()
            if served is None and fotokopi_queue.is_empty():
                print("gak ada pesanan antrian.")

        elif pilih == '3':
            front_order = fotokopi_queue.front()
            if front_order:
                print(f"pesanan pertama antrian: {front_order['order_id']} dengan {front_order['num_pages']} halaman")
            else:
                print("gak ada pesanan dalam antrian.")

        elif pilih == '4':
            rear_order = fotokopi_queue.rear()
            if rear_order:
                print(f"pesanan terakhir dalam antrian: {rear_order['order_id']} dengan {rear_order['num_pages']} halaman")
            else:
                print("gak ada pesanan dalam antrian.")

        elif pilih == '5':
            print(f"jumlah pesanan dalam antrian ada: {fotokopi_queue.size()}")

        elif pilih == '6':
            fotokopi_queue.display_queue()

        elif pilih == '0':
            print("makasie gys have fun!!")
            break

        else:
            print("pilihan gak valid, ulangi pilih lagi.")


# Menjalankan program langsung
main()
```

```python
#Soal Latihan 3: Antrian Layanan Parkir Mall
from collections import deque

class QueueList:
    def __init__(self, kapasitas_max=15):
        self.queue = deque()
        self.waiting_list = deque()
        self.kapasitas_max = kapasitas_max

    def enqueue(self, nama_kendaraan, is_listrik=False):
        # Kendaraan listrik mendapatkan prioritas
        if is_listrik:
            print(f"widih parkir prioritas! kendaraan {nama_kendaraan} listrik dapat prioritas masuk parkiran")
            return True

        # Jika parkiran belum penuh, tambahkan kendaraan ke antrian
        if len(self.queue) < self.kapasitas_max:
            self.queue.append(nama_kendaraan)
            print(f"kendaraan {nama_kendaraan} memasuki parkiran. area terisi: {len(self.queue)}/{self.kapasitas_max}")
            return True
        else:
            # Jika parkiran penuh, masukkan ke daftar tunggu
            self.waiting_list.append(nama_kendaraan)
            print(f"yahh parkiran penuh, Kendaraan {nama_kendaraan} ditambahkan ke daftar tunggu. tunggu di: {len(self.waiting_list)}")
            return False

    def dequeue(self):
        if not self.is_empty():
            exited_kendaraan = self.queue.popleft()  # Lebih efisien daripada pop(0)
            print(f"kendaraan {exited_kendaraan} keluar dari parkiran.")

            # Jika ada kendaraan di daftar tunggu, pindahkan ke parkiran
            if self.waiting_list:
                next_kendaraan = self.waiting_list.popleft()
                self.queue.append(next_kendaraan)
                print(f"kendaraan {next_kendaraan} akan dipindahkan dari daftar tunggu ke parkiran.")

            return exited_kendaraan
        return None

    def front(self):
        return self.queue[0] if not self.is_empty() else None

    def rear(self):
        return self.queue[-1] if not self.is_empty() else None

    def is_empty(self):
        return len(self.queue) == 0

    def size(self):
        return len(self.queue)

    def display_queue(self):
        if self.is_empty():
            print("hamdalah parkiran kosong.")
        else:
            print(f"kendaraan dalam parkiran ({len(self.queue)}/{self.kapasitas_max}):")
            for i, kendaraan in enumerate(self.queue, 1):
                print(f"{i}. {kendaraan}")

        if self.waiting_list:
            print(f"\nlist tunggu ({len(self.waiting_list)} kendaraan):")
            for i, kendaraan in enumerate(self.waiting_list, 1):
                print(f"{i}. {kendaraan}")

# Membuat plat nomor
def generate_plat_nomor(kota_lahir, npm):
    # Mengambil dua huruf pertama kota lahir
    kota_prefix = kota_lahir[:2].upper()
    # Mengambil dua digit terakhir NPM
    npm_suffix = npm[-2:]

    return f"{kota_prefix}{npm_suffix}"

# Program utama
def main():
    parking_queue = QueueList(kapasitas_max=15)

    while True:
        print("\nANTRIAN PARKIR MALL ")
        print("1. Kendaraan Masuk")
        print("2. Kendaraan Keluar")
        print("3. Melihat Kendaraan Pertama")
        print("4. Melihat Kendaraan Terakhir")
        print("5. Mengecek Jumlah Kendaraan dalam Parkiran")
        print("6. Menampilkan Semua Kendaraan")
        print("0. Keluar")

        pilih = input("Pilih menu (0-6): ")

        if pilih == '1':
            kota_lahir = input("Masukkan kota lahir kamu: ")
            npm = input("Masukkan NPM kamu: ")
            nama_kendaraan = generate_plat_nomor(kota_lahir, npm)

            is_listrik = input("punyamu kendaraan listrik? (y/n): ").lower() == 'y'
            parking_queue.enqueue(nama_kendaraan, is_listrik)

        elif pilih == '2':
            exited = parking_queue.dequeue()
            if exited is None:
                print("gak ada kendaraan di dalam parkiran.")

        elif pilih == '3':
            front_kendaraan = parking_queue.front()
            if front_kendaraan:
                print(f"Kendaraan pertama di dalam parkiran: {front_kendaraan}")
            else:
                print("Gak ada kendaraan di dalam parkiran.")

        elif pilih == '4':
            rear_kendaraan = parking_queue.rear()
            if rear_kendaraan:
                print(f"Kendaraan terakhir di dalam parkiran: {rear_kendaraan}")
            else:
                print("gak ada kendaraan di dalam parkiran.")

        elif pilih == '5':
            print(f"Jumlah kendaraan di dalam parkiran: {parking_queue.size()}/{parking_queue.kapasitas_max}")

        elif pilih == '6':
            parking_queue.display_queue()

        elif pilih == '0':
            print("bye guys hati hatiiii!")
            break

        else:
            print("Pilihan gak valid, coba pilih lagi.")

# Menjalankan
main()
```

```python
#Soal Latihan 4: Antrian Pengambilan Sertifikat di Kampus
from collections import deque

#Soal Latihan 4: Antrian Pengambilan Sertifikat di Kampus
class DaftarAntrian:
    def __init__(self, kapasitas_maksimal=20):
        self.antrian_reguler = deque()  # Antrian untuk mahasiswa reguler
        self.antrian_berprestasi = deque() # Antrian untuk mahasiswa berprestasi
        self.kapasitas_maksimal = kapasitas_maksimal

    def masuk_antrian(self, id_mahasiswa, peserta_kompetisi_nasional=False):
        # Periksa apakah mahasiswa adalah peserta kompetisi nasional
        if peserta_kompetisi_nasional:
            print(f"KEREN BRO priority! Mahasiswa {id_mahasiswa} (Peserta Kompetisi Nasional) dapat langsung mengambil sertifikat!")
            return True

        # Jika jumlah mahasiswa melebihi kapasitas, bagi ke dua jalur
        total_mahasiswa = len(self.antrian_reguler) + len(self.antrian_berprestasi)

        if total_mahasiswa >= self.kapasitas_maksimal:
            # Jika mahasiswa berprestasi, masukkan ke jalur mahasiswa berprestasi
            if input("Apakah mahasiswa ini berprestasi? (y/n): ").lower() == 'y':
                self.antrian_berprestasi.append(id_mahasiswa)
                print(f"Mahasiswa ini {id_mahasiswa} ditambahkan ke jalur mahasiswa berprestasi. Posisi ke berapa yaa: {len(self.antrian_berprestasi)}")
            else:
                self.antrian_reguler.append(id_mahasiswa)
                print(f"Mahasiswa ini {id_mahasiswa} ditambahkan ke jalur mahasiswa reguler. Posisi ke berapa yaaa: {len(self.antrian_reguler)}")
        else:
            # Jika belum mencapai kapasitas, masukkan ke jalur reguler
            self.antrian_reguler.append(id_mahasiswa)
            print(f"Mahasiswa ini {id_mahasiswa} ditambahkan ke antrian. Posisi berapa yaaa: {len(self.antrian_reguler)}")

        return True

    def keluar_antrian(self):
        # Prioritaskan mahasiswa berprestasi jika ada
        if len(self.antrian_berprestasi) > 0:
            mahasiswa_dilayani = self.antrian_berprestasi.popleft()
            print(f"Mahasiswa berprestasi dan lucu ini {mahasiswa_dilayani} sedang dilayani.")
            return mahasiswa_dilayani

        # Kemudian layani mahasiswa reguler
        if len(self.antrian_reguler) > 0:
            mahasiswa_dilayani = self.antrian_reguler.popleft()
            print(f"Mahasiswa {mahasiswa_dilayani} sedang dilayani.")
            return mahasiswa_dilayani

        return None

    def depan(self):
        if len(self.antrian_berprestasi) > 0:
            return self.antrian_berprestasi[0]
        elif len(self.antrian_reguler) > 0:
            return self.antrian_reguler[0]
        return None

    def belakang(self):
        if len(self.antrian_reguler) > 0:
            return self.antrian_reguler[-1]
        elif len(self.antrian_berprestasi) > 0:
            return self.antrian_berprestasi[-1]
        return None

    def kosong(self):
        return len(self.antrian_reguler) == 0 and len(self.antrian_berprestasi) == 0

    def ukuran(self):
        return len(self.antrian_reguler) + len(self.antrian_berprestasi)

    def tampilkan_antrian(self):
        if self.kosong():
            print("Antrian kosong.")
        else:
            total = len(self.antrian_reguler) + len(self.antrian_berprestasi)
            print(f"Total mahasiswa dalam antrian: {total}")

            if len(self.antrian_reguler) > 0:
                print(f"\nJalur Mahasiswa Reguler ({len(self.antrian_reguler)} mahasiswa):")
                for i, mahasiswa in enumerate(self.antrian_reguler, 1):
                    print(f"{i}. {mahasiswa}")

            if len(self.antrian_berprestasi) > 0:
                print(f"\nJalur Mahasiswa Berprestasi ({len(self.antrian_berprestasi)} mahasiswa):")
                for i, mahasiswa in enumerate(self.antrian_berprestasi, 1):
                    print(f"{i}. {mahasiswa}")


# Fungsi untuk membuat nomor antrian mahasiswa
def buat_id_mahasiswa(program_studi, npm):
    # Mengambil huruf pertama program studi
    awalan_program = program_studi[0].upper()
    # Mengambil dua digit terakhir NPM
    akhiran_npm = npm[-2:]

    return f"{awalan_program}{akhiran_npm}"


# Program utama
def main():
    antrian_sertifikat = DaftarAntrian(kapasitas_maksimal=20)

    while True:
        print("\nANTRIAN PENGAMBILAN SERTIFIKAT HEHE")
        print("1. Menambahkan Mahasiswa ke Antrian")
        print("2. Melayani Mahasiswa")
        print("3. Melihat Mahasiswa Pertama")
        print("4. Melihat Mahasiswa Terakhir")
        print("5. Mengecek Jumlah Mahasiswa dalam Antrian")
        print("6. Menampilkan Semua Antrian")
        print("0. Keluar skrg pls")

        pilihan = input("Pilih menu (0-6): ")

        if pilihan == '1':
            program_studi = input("Masukkan program studi kamu dong bub: ")
            npm = input("Masukkan NPM kamu dulu bub: ")
            id_mahasiswa = buat_id_mahasiswa(program_studi, npm)

            # Tambahan untuk VIP: cek jika NPM mengandung '7'
            if '7' in npm:
                print(f"Mahasiswa ini {id_mahasiswa} adalah VIP (NPM yang ada angka 7)!")
                peserta_nasional = True
            else:
                peserta_nasional = input("Apakah kamu pernah mengikuti kompetisi tingkat nasional bub? (y/n): ").lower() == 'y'

            antrian_sertifikat.masuk_antrian(id_mahasiswa, peserta_nasional)

        elif pilihan == '2':
            terlayani = antrian_sertifikat.keluar_antrian()
            if terlayani is None:
                print("Gak ada mahasiswa dalam antrian.")

        elif pilihan == '3':
            mahasiswa_depan = antrian_sertifikat.depan()
            if mahasiswa_depan:
                print(f"Mahasiswa pertama dalam antrian: {mahasiswa_depan}")
            else:
                print("Gak ada mahasiswa dalam antrian.")

        elif pilihan == '4':
            mahasiswa_belakang = antrian_sertifikat.belakang()
            if mahasiswa_belakang:
                print(f"Mahasiswa terakhir dalam antrian: {mahasiswa_belakang}")
            else:
                print("Gak ada mahasiswa dalam antrian.")

        elif pilihan == '5':
            print(f"Jumlah mahasiswa dalam antrian berapa yak: {antrian_sertifikat.ukuran()}")

        elif pilihan == '6':
            antrian_sertifikat.tampilkan_antrian()

        elif pilihan == '0':
            print("bye guys, selamat ya!")
            break

        else:
            print("Pilihan gak valid to. coba pilih lagi.")


# Menjalankan program langsung
main()
```
