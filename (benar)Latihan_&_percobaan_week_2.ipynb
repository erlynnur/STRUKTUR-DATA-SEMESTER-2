```python
#percobaan 1
# Definisi Node dalam Linked List
class TaskNode:
    def __init__(self, task_name):
        self.task_name = task_name #Menyimpan nama tugas
        self.next = None #Pointer ke tugas berikutnya

#Definisi Linked List untuk Daftar Tugas
class TaskList:
    def __init__(self):
        self.head = None #Awalnya tidak ada tugas (daftar kosong)

   #Menambahkan tugas ke Linked list
    def add_task(self, task_name):
       new_task = TaskNode(task_name)
       if not self.head:
           self.head = new_task #Jika list kosong, tugas pertama menjadi head

       else:
          current = self.head
          while current.next: #Mencari node terakhir dalam list
              current = current.next
          current.next = new_task #Menambahkan tugas di akhir list

    #Menampilkan daftar tugas
    def show_tasks(self):
        if not self.head:
            print("\n Tidak ada tugas hari ini!")
            return

        print("\n Daftar tugas hari ini:")
        current = self.head
        count = 1
        while current:
           print(f"{count}. {current.task_name}")
           current = current.next
           count += 1

# ==== Program Utama(main) ====#

#membuat objek daftar tugas
task_list = TaskList()

#Meminta pengguna masukkan jumlah tugas
num_tasks = int(input("Masukkan jumlah tugas hari ini:"))

#Loop untuk memasukkan tugas ke dalam linked list
for i in range(1, num_tasks + 1):
  task_name = input(f"Masukkan tugas ke-{i}: ")
  task_list.add_task(task_name)

#menampilkan hasil daftar tugas
task_list.show_tasks()
```

```python
####### percobaan 2 #######
#Definisi Node dalam Doubly Linked List
class PageNode:
  def __init__(self, url):
      self.url = url # URL halaman web
      self.next = None # Pointer ke halaman berikutnya
      self.prev = None # Pointer ke halaman sebelumnya
#Definisi Doubly Linked List untuk Riwayat Peramban
class BrowserHistory:
  def __init__(self):
      self.current = None # Pointer ke halaman yang sedang dibuka
#Menambahkan halaman baru ke dalam riwayat
  def visit_page(self, url):
      new_page = PageNode (url)
      if not self.current:
        self.current = new_page # Jika riwayat kosong, halaman ini menjadi halaman pertama
      else:
        new_page.prev = self.current
        self.current.next = new_page
        self.current = new_page # Pindahkan posisi ke halaman baru
#Menampilkan riwayat kunjungan
  def show_history(self):
      if not self.current:
          print("\n Riwayat kosong.")
          return
      print("\n Riwayat Peramban Web:")
      temp = self.current
      while temp.prev: # Kembali ke halaman pertama
        temp = temp.prev
      count = 1
      while temp:
        print(f"{count}. {temp.url}")
        temp = temp.next
        count += 1
#Navigasi mundur ke halaman sebelumnya
  def go_back(self):
      if self.current and self.current.prev:
          self.current = self.current.prev
          print(f"\n Anda kembali ke: {self.current.url}")
      else:
          print("\n Tidak bisa kembali, ini adalah halaman pertama!")
#Navigasi maju ke halaman berikutnya
  def go_forward(self):
      if self.current and self.current.next:
        self.current = self.current.next
        print (f"\n Anda maju ke: {self.current.url}")
      else:
        print("\n Tidak bisa maju, ini adalah halaman terakhir!")
#====== Program Utama (Main) ------#
#Membuat objek riwayat peramban
history = BrowserHistory()
#Meminta pengguna memasukkan jumlah halaman yang dikunjungi
num_pages = int(input("\nMasukkan jumlah halaman yang dikunjungi: "))
#Loop untuk memasukkan URL halaman ke dalam Doubly Linked List
for i in range(1, num_pages + 1):
  url = input(f"Masukkan URL halaman ke-{i}: ")
  history.visit_page(url)
#Menampilkan riwayat yang telah dimasukkan
history.show_history()
#Menu navigasi pengguna
while True:
  command = input("\nTekan 'b' untuk kembali, Tekan 'f' untuk maju, Tekan 'x' untuk keluar.\nMasukkan perintah: ").lower()

  if command == 'b':
      history.go_back()
  elif command == 'f':
      history.go_forward()
  elif command == 'x':
      print("\nKeluar dari riwayat peramban.")
      break
  else:
      print("\nPerintah tidak dikenal.")
```

```python
# latihan 2
# Definisi Node dalam Doubly Linked List untuk folder
class FolderNode:
    def __init__(self, name):
        self.name = name  # Nama folder
        self.next = None  # Pointer ke folder berikutnya
        self.prev = None  # Pointer ke folder sebelumnya

# Definisi Doubly Linked List untuk Riwayat Navigasi Folder
class FolderNavigationHistory:
    def __init__(self):
        self.current = None  # Pointer ke folder yang sedang dibuka

    # Menambahkan folder baru ke dalam riwayat
    def navigate_to_folder(self, folder_name):
        new_folder = FolderNode(folder_name)

        if not self.current:
            self.current = new_folder  # Jika riwayat kosong, folder ini menjadi folder pertama
        else:
            # Jika kita navigasi setelah mundur, hapus semua folder setelah current
            if self.current.next:
                temp = self.current.next
                while temp:
                    next_folder = temp.next
                    temp.next = None
                    temp.prev = None
                    temp = next_folder

            # Tambahkan folder baru setelah current
            new_folder.prev = self.current
            self.current.next = new_folder
            self.current = new_folder  # Pindahkan posisi ke folder baru

        print(f"\nAnda bernavigasi ke: {folder_name}")

    # Menampilkan riwayat navigasi folder
    def show_history(self):
        if not self.current:
            print("\nRiwayat navigasi kosong.")
            return

        print("\nRiwayat Navigasi File Explorer:")

        # Kembali ke folder pertama
        temp = self.current
        while temp.prev:
            temp = temp.prev

        # Tampilkan semua folder dalam riwayat
        count = 1
        while temp:
            if temp == self.current:
                print(f"{count}. {temp.name} (Folder Saat Ini)")
            else:
                print(f"{count}. {temp.name}")
            temp = temp.next
            count += 1

    # Navigasi mundur ke folder sebelumnya
    def go_back(self):
        if self.current and self.current.prev:
            self.current = self.current.prev
            print(f"\nAnda kembali ke: {self.current.name}")
        else:
            print("\nTidak bisa kembali, ini adalah folder pertama!")

    # Navigasi maju ke folder berikutnya
    def go_forward(self):
        if self.current and self.current.next:
            self.current = self.current.next
            print(f"\nAnda maju ke: {self.current.name}")
        else:
            print("\nTidak bisa maju, ini adalah folder terakhir!")

# ====== Program Utama (Main) ======
# Membuat objek riwayat navigasi folder
history = FolderNavigationHistory()

# Meminta pengguna memasukkan digit terakhir NPM
last_digit_npm = int(input("\nMasukkan digit terakhir NPM Anda: "))
num_folders = last_digit_npm  # Jumlah folder yang akan dinavigasi sesuai digit terakhir NPM

# Folder awal (root)
history.navigate_to_folder("C:\\")

# Loop untuk memasukkan nama folder ke dalam Doubly Linked List
for i in range(1, num_folders + 1):
    folder_name = input(f"Masukkan nama folder ke-{i}: ")
    history.navigate_to_folder(folder_name)

# Menampilkan riwayat yang telah dimasukkan
history.show_history()

# Menu navigasi pengguna
while True:
    command = input("\nTekan 'b' untuk kembali, Tekan 'f' untuk maju, Tekan 'n' untuk folder baru, Tekan 'x' untuk keluar.\nMasukkan perintah: ").lower()

    if command == 'b':
        history.go_back()
        history.show_history()
    elif command == 'f':
        history.go_forward()
        history.show_history()
    elif command == 'n':
        new_folder = input("Masukkan nama folder baru: ")
        history.navigate_to_folder(new_folder)
        history.show_history()
    elif command == 'x':
        print("\nKeluar dari riwayat navigasi folder.")
        break
    else:
        print("\nPerintah tidak dikenal.")
```

```python
#latihan 3
#operasi komplit
# Definisi Node dalam Singly Linked List
class Node:
    def __init__(self, data):
        self.data = data  # Menyimpan nilai node
        self.next = None  # Pointer ke node berikutnya

# Definisi Singly Linked List dengan Berbagai Operasi
class LinkedList:
    def __init__(self):
        self.head = None  # Awalnya, daftar kosong

    # Menambahkan elemen di awal
    def add_at_beginning(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    # Menambahkan elemen di akhir
    def add_at_end(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node

    # Menambahkan elemen di posisi tertentu
    def add_at_position(self, position, data):
        if position < 0:
            print("Posisi tidak valid!")
            return
        new_node = Node(data)
        if position == 0:
            self.add_at_beginning(data)
            return
        temp = self.head
        for _ in range(position - 1):
            if not temp:
                print(" Posisi melebihi panjang daftar!")
                return
            temp = temp.next
        if not temp:
            print("Posisi melebihi panjang daftar!")
            return
        new_node.next = temp.next
        temp.next = new_node

    # Menghapus elemen di awal
    def delete_at_beginning(self):
        if not self.head:
            print("Daftar kosong, tidak ada yang bisa dihapus!")
            return
        self.head = self.head.next
        return

    # Menghapus elemen di akhir
    def delete_at_end(self):
        if not self.head:
            print(" Daftar kosong, tidak ada yang bisa dihapus!")
            return
        if not self.head.next:
            self.head = None
            return
        temp = self.head
        while temp.next.next:
            temp = temp.next
        temp.next = None

    # Menghapus elemen di posisi tertentu
    def delete_at_position(self, position):
        if not self.head:
            print(" Daftar kosong, tidak ada yang bisa dihapus!")
            return
        if position == 0:
            self.head = self.head.next
            return
        temp = self.head
        for _ in range(position - 1):
            if not temp or not temp.next:
                print(" Posisi tidak ditemukan!")
                return
            temp = temp.next
        temp.next = temp.next.next

    # Mencari elemen dalam Linked List
    def search(self, key):
        temp = self.head
        while temp:
            if temp.data == key:
                print( f" Elemen {key} ditemukan dalam daftar!")
                return
            temp = temp.next
        print( f" Elemen {key} tidak ditemukan dalam daftar.")

    # Membalik (reverse) Linked List
    def reverse(self):
        prev = None
        current = self.head
        while current:
            next_node = current.next
            current.next = prev
            prev = current
            current = next_node
        self.head = prev

    # Menampilkan Linked List
    def display(self):
        if not self.head:
            print(" Daftar kosong!")
            return
        temp = self.head
        while temp:
            print(temp.data, end="-> ")
            temp = temp.next
        print("None")

# ======== Program Utama (Main) ========

# Membuat objek Linked List
ll = LinkedList()

# Meminta pengguna memasukkan elemen awal
num_elements = int(input("\nMasukkan jumlah awal elemen dalam daftar: "))
for i in range(num_elements):
    data = int(input(f"Masukkan data ke-{i+1}: "))
    ll.add_at_end(data)

# Menu Operasi Linked List
while True:
    print("\n Pilih Operasi:")
    print("1. Tambah elemen di awal")
    print("2. Tambah elemen di akhir")
    print("3. Tambah elemen di posisi tertentu")
    print("4. Hapus elemen di awal")
    print("5. Hapus elemen di akhir")
    print("6. Hapus elemen di posisi tertentu")
    print("7. Cari elemen dalam daftar")
    print("8. Balikkan daftar")
    print("9. Tampilkan daftar")
    print("0. Keluar")

    choice = int(input("\nMasukkan pilihan: "))

    if choice == 1:
        data = int(input("Masukkan data yang ingin ditambahkan di awal: "))
        ll.add_at_beginning(data)
    elif choice == 2:
        data = int(input("Masukkan data yang ingin ditambahkan di akhir: "))
        ll.add_at_end(data)
    elif choice == 3:
        position = int(input("Masukkan posisi untuk menambahkan data: "))
        data = int(input("Masukkan data yang ingin ditambahkan: "))
        ll.add_at_position(position, data)
    elif choice == 4:
        ll.delete_at_beginning()
    elif choice == 5:
        ll.delete_at_end()
    elif choice == 6:
        position = int(input("Masukkan posisi elemen yang ingin dihapus: "))
        ll.delete_at_position(position)
    elif choice == 7:
        key = int(input("Masukkan elemen yang ingin dicari: "))
        ll.search(key)
    elif choice == 8:
        ll.reverse()
        print("Daftar berhasil dibalik!")
    elif choice == 9:
        ll.display()
    elif choice == 0:
        print("\n Keluar dari program.")
        break
    else:
        print("\n Pilihan tidak valid, silakan coba lagi!")
```

```python
# latihan 3
# Definisi Node dalam Doubly Linked List
class FolderNode:
    def __init__(self, folder_nama):
        self.folder_nama = folder_nama  # Nama folder
        self.next = None  # Pointer ke folder berikutnya
        self.prev = None  # Pointer ke folder sebelumnya

# Definisi Doubly Linked List untuk Riwayat File Explorer
class FileExplorerHistory:
    def __init__(self):
        self.current = None  # Pointer ke folder yang sedang dibuka

    # Menambahkan folder baru ke dalam riwayat
    def visit_folder(self, folder_nama):
        new_folder = FolderNode(folder_nama)
        if not self.current:
            self.current = new_folder  # Jika riwayat kosong, folder ini menjadi folder pertama
        else:
            new_folder.prev = self.current
            self.current.next = new_folder
            self.current = new_folder  # Pindahkan posisi ke folder baru

    # Menampilkan riwayat kunjungan folder
    def show_history(self):
        if not self.current:
            print("\nRiwayat kosong.")
            return

        print("\nRiwayat File Explorer:")
        temp = self.current
        while temp.prev:  # Kembali ke folder pertama
            temp = temp.prev

        count = 1
        while temp:
            print(f"{count}. {temp.folder_nama}")
            temp = temp.next
            count += 1

    # Navigasi mundur ke folder sebelumnya
    def go_back(self):
        if self.current and self.current.prev:
            self.current = self.current.prev
            print(f"\nAnda kembali ke: {self.current.folder_nama}")
        else:
            print("\nTidak bisa kembali, ini adalah folder pertama!")

    # Navigasi maju ke folder berikutnya
    def go_forward(self):
        if self.current and self.current.next:
            self.current = self.current.next
            print(f"\nAnda maju ke: {self.current.folder_nama}")
        else:
            print("\nTidak bisa maju, ini adalah folder terakhir!")

# ====== Program Utama (Main) ======
# Membuat objek riwayat file explorer
history = FileExplorerHistory()

# Meminta pengguna memasukkan jumlah folder yang dikunjungi sesuai digit terakhir NPM
npm_last_digit = int(input("\nMasukkan digit terakhir NPM Anda: "))
num_folders = npm_last_digit

print(f"Anda akan memasukkan {num_folders} folder untuk navigasi.")

# Loop untuk memasukkan nama folder ke dalam Doubly Linked List
for i in range(1, num_folders + 1):
    folder_nama = input(f"Masukkan nama folder ke-{i}: ")
    history.visit_folder(folder_nama)

# Menampilkan riwayat yang telah dimasukkan
history.show_history()

# Menu navigasi pengguna
while True:
    command = input("\nTekan 'b' untuk kembali, Tekan 'f' untuk maju, Tekan 'x' untuk keluar.\nMasukkan perintah: ").lower()

    if command == 'b':
        history.go_back()
    elif command == 'f':
        history.go_forward()
    elif command == 'x':
        print("\nKeluar dari riwayat File Explorer.")
        break
    else:
        print("\nPerintah tidak dikenal.")
```
