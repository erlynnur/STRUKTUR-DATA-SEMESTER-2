```python
import random
import math

# Fungsi Linear Search
def linear_search(arr, target):
    # Variabel untuk menghitung perbandingan
    comparisons = 0

    # Loop melalui semua elemen array
    for i in range(len(arr)):
        comparisons += 1  # Tambahkan jumlah perbandingan
        if arr[i] == target:  # Jika elemen ditemukan
            return i, comparisons  # Kembalikan indeks dan jumlah perbandingan

    # Jika tidak ditemukan, kembalikan -1
    return -1, comparisons

# Fungsi Binary Search
def binary_search(arr, target):
    # Inisialisasi variabel awal
    low = 0
    high = len(arr) - 1
    comparisons = 0  # Variabel untuk menghitung perbandingan

    # Loop selama low <= high
    while low <= high:
        mid = (low + high) // 2  # Hitung indeks tengah
        comparisons += 1  # Tambahkan jumlah perbandingan

        if arr[mid] == target:  # Jika elemen ditemukan
            return mid, comparisons  # Kembalikan indeks dan jumlah perbandingan
        elif arr[mid] < target:  # Jika target lebih besar, cari di kanan
            low = mid + 1
        else:  # Jika target lebih kecil, cari di kiri
            high = mid - 1

    # Jika tidak ditemukan, kembalikan -1
    return -1, comparisons

# Fungsi Jump Search
def jump_search(arr, target):
    # Inisialisasi variabel awal
    n = len(arr)
    step = int(math.sqrt(n))  # Hitung ukuran lompatan
    prev = 0
    comparisons = 0  # Variabel untuk menghitung perbandingan

    # Melompat ke blok berikutnya
    while prev < n and arr[min(step, n) - 1] < target:
        comparisons += 1  # Tambahkan jumlah perbandingan
        prev = step
        step += int(math.sqrt(n))
        if prev >= n:  # Jika melewati batas array
            return -1, comparisons

    # Pencarian linier di dalam blok
    for i in range(prev, min(step, n)):
        comparisons += 1  # Tambahkan jumlah perbandingan
        if arr[i] == target:  # Jika elemen ditemukan
            return i, comparisons  # Kembalikan indeks dan jumlah perbandingan

    # Jika tidak ditemukan, kembalikan -1
    return -1, comparisons

# Membuat daftar angka acak
random_numbers = random.sample(range(1, 1001), 100)

# Mengurutkan angka untuk Binary Search dan Jump Search
sorted_numbers = sorted(random_numbers)

# Menentukan target pencarian (elemen tengah dari array terurut)
middle_index = len(sorted_numbers) // 2
target = sorted_numbers[middle_index]

# Menampilkan informasi
print("PERBANDINGAN ALGORITMA PENCARIAN")
print(f"Daftar berisi {len(sorted_numbers)} angka acak dalam rentang 1-1000")
print(f"Target pencarian: {target} (elemen tengah dari daftar terurut)")

# Eksekusi algoritma pencarian
print("\nHasil Pencarian:")

# a. Linear Search
linear_index, linear_comparisons = linear_search(sorted_numbers, target)
print(f"Linear Search: Ditemukan pada indeks {linear_index}, total perbandingan: {linear_comparisons}")

# b. Binary Search
binary_index, binary_comparisons = binary_search(sorted_numbers, target)
print(f"Binary Search: Ditemukan pada indeks {binary_index}, total perbandingan: {binary_comparisons}")

# c. Jump Search
jump_index, jump_comparisons = jump_search(sorted_numbers, target)
print(f"Jump Search: Ditemukan pada indeks {jump_index}, total perbandingan: {jump_comparisons}")

# Perbandingan
print("\n PERBANDINGAN")
print("-" * 70)
print(f"{'Algoritma':<20} | {'Jumlah Perbandingan':<20} | {'Indeks Target':<15}")
print("-" * 70)
print(f"{'Linear Search':<20} | {linear_comparisons:<20} | {linear_index:<15}")
print(f"{'Binary Search':<20} | {binary_comparisons:<20} | {binary_index:<15}")
print(f"{'Jump Search':<20} | {jump_comparisons:<20} | {jump_index:<15}")
print("-" * 70)

# Menentukan algoritma paling efisien
min_comparisons = min(linear_comparisons, binary_comparisons, jump_comparisons)
if min_comparisons == linear_comparisons:
    most_efficient = "Linear Search"
elif min_comparisons == binary_comparisons:
    most_efficient = "Binary Search"
else:
    most_efficient = "Jump Search"

print(f"\nAlgoritma terbaik berdasarkan jumlah perbandingan: {most_efficient} ({min_comparisons} perbandingan)")

```
